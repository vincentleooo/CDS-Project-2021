```python
import pandas as pd

temp_csv = "HadCRUT Temperature Anomaly.csv"
rain_csv = "rainfall.csv"
edu_csv = "EdStatsData.csv"

temperature = pd.read_csv(temp_csv)
rainfall = pd.read_csv(rain_csv)
education = pd.read_csv(edu_csv)
```

```python
print("temperature: ")
print(temperature.head(10))
print(temperature.columns)
print(temperature.size)
print()
print("rainfall: ")
print(rainfall.head(10))
print(rainfall.columns)
print(rainfall.size)
print()
print("education: ")
print(education.head(10))
print(education.columns)
print(education.size)
print()
```

```python
params = ["Country","Year","Surface temperature anomaly","Surface temperature anomaly weighted by population","Surface temperature anomaly weighted by area","Rainfall","Enrolment ratio"]
integrated_data = pd.DataFrame(columns = params)
for _, row in rainfall.iterrows():
  # rainfall
  country = row["Area"]
  year = row["Year"]
  rainfall_value = row["Value"]

  # education
  edu = education.loc[education['Country Name'] == country]
  if edu.size == 0:
    continue
  else:
    try: 
      enrolment_ratio = edu[str(year)].values[0]
      if pd.isna(enrolment_ratio):
        continue
    except:
      continue

    # temperature
    temp = temperature.loc[(temperature['Entity'] == country) & (temperature["Year"] == year)]
    if temp.size == 0:
      continue
    else:
      surface_temp = temp["Surface temperature anomaly"].values[0]
      surface_temp_population = temp["Surface temperature anomaly weighted by population"].values[0]
      surface_temp_area = temp["Surface temperature anomaly weighted by area"].values[0]

  df = pd.DataFrame([[country,year,surface_temp,surface_temp_population,surface_temp_area,rainfall_value,enrolment_ratio]],columns=params)
  integrated_data = integrated_data.append(df,ignore_index=True)
```
